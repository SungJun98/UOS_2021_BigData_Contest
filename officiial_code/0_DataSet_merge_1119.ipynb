## 요소수 중점 유통 주유소 재고현황 

```python
os.chdir('/content/drive/My Drive/contest/data')
os.listdir('./')
```

```python
a=pd.read_csv('공공데이터활용지원센터_요소수 중점 유통 주유소 재고현황_20211117.csv',encoding='CP949')
a=a[['주소','입고량','재고량']]
a
```

```python
for i in range(len(a)):
  a['입고량'][i]=a['입고량'][i].replace(',','')
  a['재고량'][i]=a['재고량'][i].replace(',','')
  a['입고량'][i]=int(a['입고량'][i])
  a['재고량'][i]=int(a['재고량'][i])
a
    
```

```python
for i in range(len(a)):
  a['주소'][i]=a['주소'][i].split()[0]
b=a.groupby('주소').sum()
b
```

```python
#입고량과 재고량을 지역에 맞게 더 함 (순서는 수도권, 강원, 대전충청, 전북, 대구경북, 광주전남,부산경남)
storage_fluid=[87200,3500,38000,16000,46000,0,47990]
spare_fluid=[59056,60,64120,12874,36823,17492,111385]
```

```python
# 입고량
storage_fluid=pd.DataFrame(storage_fluid,index=region,columns=['storage_fluid'])
storage_fluid
```

```python
# 재고량
spare_fluid=pd.DataFrame(spare_fluid,index=region,columns=['spare_fluid'])
spare_fluid
```
